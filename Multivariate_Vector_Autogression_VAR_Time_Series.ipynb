```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import sys
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from sklearn import preprocessing

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
# IMPORT DAILY CASE DATA
str_file = 'daily.csv'
df_cases = pd.read_csv(str_file)
df_cases.head()
```

```python
# FILTER DAILY CASE DATA FOR ONLY NECESSARY COLUMNS
lst_columns_to_keep = ['date', 'positiveIncrease']
df_cases = df_cases[lst_columns_to_keep]
df_cases.head()
```

```python
# MAKE DATE DATETIME
df_cases["date"] = pd.to_datetime(df_cases["date"], format='%Y%m%d')
df_cases.head()
```

```python
# SUM DATA FOR ALL OF USA
df_cases = df_cases.groupby(['date'])['positiveIncrease'].sum()
df_cases.tail()
```

```python
# IMPORT WEATHER DATA
df_weather = pd.read_excel('external_datasets.xlsx', sheet_name='average_US_temperatures_2019')
df_weather["date"] = pd.to_datetime(df_weather["date"], format='%Y%m%d')
df_weather.head()
```

```python
# IMPORT BORDER CLOSING DATA
df_border = pd.read_excel('external_datasets.xlsx', sheet_name='border_closing_2020')
df_border["date"] = pd.to_datetime(df_border["date"], format='%Y%m%d')
df_border.head()
```

```python
# IMPORT NUMBER OF TOURISTS
df_tourists = pd.read_excel('external_datasets.xlsx', sheet_name='tourists_2020')
df_tourists["date"] = pd.to_datetime(df_tourists["date"], format='%Y%m%d')
df_tourists.head()
```

```python
# MAKE A LIST OF ALL DATES TO FORECAST
from datetime import date, timedelta

date_start = date(2020, 6, 8)
date_end = date(2020, 11, 30)

int_days_in_between = date_end - date_start

lst_dates = []

for i in range(int_days_in_between.days + 1):
    day = date_start + timedelta(days=i)
    lst_dates.append(str(day))
```

```python
# MODEL
# manual loop to make the model multi-step
for date in lst_dates:
    # merge with weather data 
    df = pd.merge(
        df_cases,
        df_weather,
        left_on=["date"],
        right_on=["date"],
        how="left",
    )
    # merge with border closing data 
    df = pd.merge(
        df,
        df_border,
        left_on=["date"],
        right_on=["date"],
        how="left",
    )
    
    # error in the model with this data which is why it is commented out 
    # merge with tourist data 
#     df = pd.merge(
#         df,
#         df_tourists,
#         left_on=["date"],
#         right_on=["date"],
#         how="left",
#     )
    
    # normalize the data
    df = df.set_index('date')
    x = df[['positiveIncrease', 'border_closed']] #, 'tourists']].values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    df_scaled = min_max_scaler.fit_transform(x)
    
    # fitting the model
    model = VAR(df_scaled)#, freq='D')
    results = model.fit(maxlags=15, ic='aic') # maxlag of 15 is an arbitrary number

    # forecasting
    lag_order = results.k_ar
#     forecast = results.forecast(df.values[-lag_order:], 1) # lag_order is the recommended lag from the model but output is not as satisfactory
    forecast = results.forecast(df_scaled[-15:], 1) # 15 is arbitrary, would need to do further tinkering to find optimal value
    
    # add the previous forecast to the dataset as a new observation
    df_cases.loc[date] = round(max(min_max_scaler.inverse_transform(forecast)[0][0], 0))
    df_cases.index = pd.to_datetime(df_cases.index)

df_cases.tail(10) 
```
